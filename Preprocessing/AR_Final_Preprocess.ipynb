```
datafile = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Activity Recognition/raw_field_acc_user18.csv')
datafile = datafile.sort_values(by='datetime')
datafile
```

```
labelfile = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Activity Recognition/user18_label.csv')
labelfile = labelfile.sort_values(by='start')
labelfile
```

```
labelfile['start']= pd.to_datetime(labelfile['start'])
```

```
labelfile['finish']= pd.to_datetime(labelfile['finish'])
```

```
datafile['datetime']= pd.to_datetime(datafile['datetime'])
```

```
datafile = datafile.reset_index(drop=True)
labelfile = labelfile.reset_index(drop=True)
```

```
lidx = 0
for ind, row in datafile.iterrows():
  print(ind)
  if labelfile.loc[lidx, 'finish'] < datafile.loc[ind,'datetime']:
    lidx = lidx + 1
    if lidx == len(labelfile): 
      break
  if labelfile.loc[lidx,'start']<=datafile.loc[ind,'datetime']<= labelfile.loc[lidx,'finish']:
    datafile.loc[ind, 'label'] =  labelfile.loc[lidx, 'activity_type_id']
```
